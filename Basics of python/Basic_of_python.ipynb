Exercise 1: Prime Numbers

Write a Python program that checks whether a given number is prime or not. A prime number is a natural number greater than 1 that has no positive divisors other than 1 and itself.

```python
try:
    n = int(input("Enter a number: "))

    if n <= 1:
        print("Not Prime")
    else:
        for i in range(2, int(n**0.5) + 1):
            if n % i == 0:
                print("Not Prime")
                break
        else:
            print("Prime")

except ValueError:
    print("Invalid entry, please provide a number")
```

Exercise 3: Squares of Even/Odd Numbers

Create a Python script that prints the squares of all even or odd numbers within the range of 100 to 200. Choose either even or odd numbers and document your choice in the code.

```python
# Squares of Even or Odd Numbers based on user choice

choice = input("Enter 'even' or 'odd': ").strip().lower()

for num in range(100, 201):
    if choice == "even" and num % 2 == 0:
        print(f"{num}^2 = {num**2}")
    elif choice == "odd" and num % 2 != 0:
        print(f"{num}^2 = {num**2}")
```

Exercise 4: Word counter

write a program to count the number of words in a given text.

```python
# Taking input from user
input_text = input("Enter the text: ")
```

```python
words = input_text.split()
```

```python
word_count = {} # # Dictionary to store word counts

for word in words:
    word_count[word] = word_count.get(word, 0) + 1 # store count of word in word_count

# Display result
for word, count in word_count.items():
    print(f"'{word}': {count}")
```

Exercise 5: Check for Palindrome

Write a Python function called is_palindrome that takes a string as input and returns True if the string is a palindrome, and False otherwise. A palindrome is a word, phrase, number, or other sequence of characters that reads the same forward and backward, ignoring spaces, punctuation, and capitalization.

```python
def is_palindrome(text: str) -> bool:
    cleaned = ''.join(ch.lower() for ch in text if ch.isalnum()) # convert to lower case and Ignores spaces, punctuation, symbols
    return cleaned == cleaned[::-1] # Check palindrome
```

```python
user_text = input("Enter a string to check if it's a palindrome: ") # user Input

if is_palindrome(user_text):
    print("True! It's a palindrome")
else:
    print("False! It's not a palindrome")
```
