```python
import warnings
warnings.filterwarnings('ignore')
from collections import Counter

# Import necessary packages
import pandas as pd
import numpy as np

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load dataset
df = pd.read_csv("Alphabets_data.csv")
```

```python
# Display first 5 rows
df.head()
```

```python
# Cell 3: Basic exploration of dataset
print("Shape of dataset:", df.shape)
print("\nDataset info:\n")
print(df.info())
```

```python
print("\nChecking missing values:\n")
print(df.isnull().sum())
```

```python
print("\nTarget column unique classes:\n")
print(df['letter'].unique())
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
# Split features (X) and target (y)
X = df.drop("letter", axis=1)   # Features
y = df["letter"]                # Target labels
```

```python
# Encode target labels into numbers
encoder = LabelEncoder()
y_encoded = encoder.fit_transform(y)
```

```python
print("Encoded labels example:", y_encoded[:10])
```

```python
# Normalize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
print("Scaled training sample:\n", X_scaled[0])
```

```python
# Split the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_encoded, test_size=0.2, random_state=42)

print('x_train',X_train.shape)
print('x_test',X_test.shape)
print('y_train',y_train.shape)
print('y_test',y_test.shape)
```

#ANN Model Building

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam

from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import GridSearchCV

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
# Build a basic ANN model
model = Sequential()
model.add(Dense(64, input_dim=X_train.shape[1], activation='relu'))  # Hidden layer
model.add(Dense(32, activation='relu'))  # Another hidden layer
model.add(Dense(len(encoder.classes_), activation='softmax'))  # Output layer for multi-class classification

# Compile the model
model.compile(optimizer=Adam(learning_rate=0.001), loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test))

# Predict on test data
y_pred = np.argmax(model.predict(X_test), axis=-1)
```

```python
# Evaluation metrics
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
print(classification_report(y_test, y_pred, target_names=encoder.classes_))
```

```python
print("\nClassification Report:\n")
print(classification_report(y_test, y_pred))

print("\nConfusion Matrix:\n")
print(confusion_matrix(y_test, y_pred))
```

Hyperparameter Tuning (Manual)

```python
# Function to create model for GridSearch
def create_model(learning_rate=0.001, activation="relu"):
    model = Sequential()
    model.add(Dense(64, activation=activation, input_shape=(X_train.shape[1],)))
    model.add(Dense(32, activation=activation))
    model.add(Dense(len(encoder.classes_), activation="softmax"))

    model.compile(
        optimizer=Adam(learning_rate=learning_rate),
        loss="sparse_categorical_crossentropy",
        metrics=["accuracy"]
    )
    return model
```

```python
# Wrap the model for scikit-learn compatibility
model = KerasClassifier(model=create_model, verbose=0)
model
```

 # Hyperparameter Tuning using GridSearchCV

```python
# Define hyperparameters to tune
param_grid = {
    'batch_size': [32, 64],
    'epochs': [10, 20],
    'model__learning_rate': [0.001, 0.01], # Updated parameter name
    'model__activation': ['relu', 'tanh'] # Updated parameter name
}

# Perform GridSearchCV
grid = GridSearchCV(estimator=model, param_grid=param_grid, cv=3)
grid_result = grid.fit(X_train, y_train)
grid_result
```

```python
# Display best parameters and score
print(f'Best parameters: {grid_result.best_params_}')
print(f'Best score: {grid_result.best_score_}')
```

#Evaluation

```python
# Evaluate the best model on the test set
best_model = grid_result.best_estimator_
y_pred_best = best_model.predict(X_test)
accuracy_best = accuracy_score(y_test, y_pred_best)
print(f'Accuracy of the best model: {accuracy_best}')
print(classification_report(y_test, y_pred_best, target_names=encoder.classes_))

# Compare performance with the basic model
print(f'Accuracy of the basic model: {accuracy}')

```

#Effects of Hyperparameter Tuning:

The GridSearchCV procedure systematically explored combinations of hyperparameters such as learning rate, batch size, number of epochs, and activation functions to identify an optimal configuration.

However, the tuned model produced the same accuracy as the baseline model, indicating that hyperparameter adjustments did not yield measurable performance gains for this dataset or architecture.

This outcome suggests that the default configuration was already near-optimal, or that model performance may be constrained more by data characteristics or model capacity than by parameter settings.

Despite no improvement in accuracy, tuning remains valuable because it validates model robustness and confirms that alternative configurations do not significantly outperform the baseline.

Hyperparameters like learning rate still influence convergence behavior, and batch size and epochs affect training efficiency, even when final accuracy remains unchanged.

Activation function selection also impacts representational capacity, though in this case it did not translate into observable metric differences.

#Conclusion:
Hyperparameter tuning remains an important step in ANN development, even when performance improvements are not realized. In this study, tuning verified that the baseline model was already performing near its achievable accuracy under the tested configurations. Future improvements may require architectural changes, feature engineering, or expanded hyperparameter search ranges rather than incremental tuning alone.
